```python
import pandas as pd

# Create the dataset
data = {
    'Country': ['USA', 'UK', 'Canada', 'Australia'],
    'Tuition': [27000, 24000, 20000, 28000],
    'LivingCost': [14000, 12000, 10000, 13000]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Add total cost column
df['TotalCost'] = df['Tuition'] + df['LivingCost']

# Show the dataset
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define custom colors again
colors = ['blue', 'goldenrod', 'green', 'red']


# Plot with hue and no legend
plt.figure(figsize=(8, 5))
sns.barplot(x='Country', y='TotalCost', hue='Country', data=df, palette=colors, legend=False, dodge=False)

plt.title("Total Education Cost by Country")
plt.xlabel("Country")
plt.ylabel("Cost (USD)")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data values
countries = df['Country']
tuition = df['Tuition']
living = df['LivingCost']

# Matching country colors (like your first chart)
colors = ['blue', 'goldenrod', 'green', 'red']

# Plot setup
plt.figure(figsize=(8, 5))

# Plot Tuition bars (bottom)
plt.bar(countries, tuition, color=colors, label='Tuition')

# Plot LivingCost bars (on top of Tuition)
plt.bar(countries, living, bottom=tuition, color=colors, alpha=0.4, label='LivingCost')

# Labels and title
plt.title("Tuition vs Living Cost by Country")
plt.xlabel("Country")
plt.ylabel("Cost (USD)")
plt.legend()
plt.tight_layout()
plt.show()
```

### 📊 Summary of Insights

- **USA** remains the most expensive country, with a total annual cost of $45,000.
- **Canada** continues to be the most affordable option at $30,000/year.
- **Australia** is the second most expensive ($40,000), followed by the **UK** at $37,000.
- The biggest difference between countries lies in tuition fees, while living expenses remain somewhat similar.

### 💬 Conclusion

This analysis helps international students better understand and compare the financial burden of studying in each country. Making an informed choice based on both tuition and living expenses can lead to significant savings.
